# Model Training

```python
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu')) # 19 for number of features
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
# No increase in validation set
# Could have kept training because validation loss is still going down
# but flattens out considerably so it wouldn't make a difference anyway
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions)) # RMSE
```

```python
mean_absolute_error(y_test,predictions)
```

```python
df['price'].describe()
```

```python
# off by about 20%
```

```python
# How much variance is being explained by your model, 1.0 is best
explained_variance_score(y_test,predictions)
```
